# Drop missing values from the columns

The function dropna() is a Pandas function that is used to remove rows or columns that contain missing values from a DataFrame. The axis parameter specifies which axis to drop the missing values from. If axis=0, then the function will drop rows that contain missing values. If axis=1, then the function will drop columns that contain missing values.

## dropna() pands function to drop none available data
### Axis 1 is for columns, Axis 0 is for rows

## fillna() pands function to fill none available data

Similar to dropna() function in pandas, we will use fillna() function in pandas to fill the none avaliable data.



there are several other methods that can be used with the fillna() function. Some examples include: <br> 

'ffill': forward fill, which means that missing values will be replaced with the previous non-null value in the same column. <br>
'backfill': backfill, which means that missing values will be replaced with the next non-null value in the same column.
'pad': short for padding, which is equivalent to ffill. <br>
'nearest': fill missing values using the nearest non-null value. <br>
'interpolate': interpolate missing values using linear interpolation. <br>
'mean': fill missing values with the mean of the non-null values in the same column. <br>
'median': fill missing values with the median of the non-null values in the same column. <br>
'mode': fill missing values with the mode (most common value) of the non-null values in the same column. <br>
